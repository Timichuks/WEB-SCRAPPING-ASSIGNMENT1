```python
#request to scrape page
page=requests.get("https://www.icc-cricket.com")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
#multiple scrapping of headers ODI women

countries = soup.find_all("span",class_='u-hide-phablet')

countries = []

for i in countries:

    countries.append(i.text)
countries

```

```python
#multiple scrapping of headers ODI women

matches = soup.find_all("td",class_='rankings-block__banner--matches')

matches = []

for i in matches:

    matches.append(i.text)
matches

```

```python
#multiple scrapping of headers ODI women

points= soup.find_all("td",class_='rankings-block__banner--points')

points=[]

for i in points:
    
    points.append(i.text)
points
    
```

```python
#multiple scrapping of headers ODI women

rating= soup.find_all("td",class_='rankings-block__banner--rating u-text-right')

rating=[]

for i in rating:
    
    rating.append(i.text)
rating
```
